```python
import fitz  # PyMuPDF
from bs4 import BeautifulSoup

def extract_html_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    html_content = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        html_content += page.get_text("html")
    return html_content

def classify_html_content(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    classifications = []

    for tag in soup.find_all(True):  # Find all tags
        if tag.name.startswith('h'):
            # Heuristic for headers (h1, h2, h3, etc.)
            classifications.append((tag.get_text(strip=True), 'Header'))
        elif 'font-weight:bold' in str(tag.get('style', '')):
            # Heuristic for titles (bold text)
            classifications.append((tag.get_text(strip=True), 'Title'))
        else:
            # Default to sentence
            classifications.append((tag.get_text(strip=True), 'Sentence'))
    
    return classifications

# Specify the path to your PDF file
pdf_path = pdfile

# Extract HTML content from the PDF
html_content = extract_html_from_pdf(pdf_path)


# Classify the HTML content
classified_content = classify_html_content(html_content)

# Print the classified content
for text, classification in classified_content:
    print(f"{classification}: {text}")

```
