This code is to verify the table created and colums.

```python
import psycopg2
import os
from dotenv import load_dotenv


# Load environment variables from the .env file
load_dotenv()

# Function to fetch table names and column information
def fetch_table_and_column_info():
    # Database connection parameters
    connection_params = {
        'host': os.getenv('DB_HOST'),
        'database': os.getenv('DB_NAME'),
        'user': os.getenv('DB_USER'),
        'password': os.getenv('DB_PASSWORD')
    }

    try:
        # Establish a database connection
        connection = psycopg2.connect(**connection_params)

        # Fetch and print table names
        with connection.cursor() as cursor:
            table_query = """
            SELECT table_name
            FROM information_schema.tables
            WHERE table_schema = 'public';
            """
            cursor.execute(table_query)
            tables = cursor.fetchall()
            print("Tables in the public schema:")
            for table in tables:
                print(f"Table: {table[0]}")

        # Fetch and print column information for 'embeddings_table'
        with connection.cursor() as cursor:
            column_query = """
            SELECT column_name, data_type 
            FROM information_schema.columns 
            WHERE table_name = 'embeddings_table';
            """
            cursor.execute(column_query)
            columns = cursor.fetchall()
            print("\nColumns in the 'embeddings_table':")
            for column in columns:
                print(f"Column: {column[0]}, Data Type: {column[1]}")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        # Close the connection
        if connection:
            connection.close()

# Call the function to fetch table and column information
fetch_table_and_column_info()
```
